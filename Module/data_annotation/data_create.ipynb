```python

def split_paragraphs_into_dataframe(large_string, dataframe, start_index):
    """
    Splits a large string into paragraphs and inserts them into the DataFrame starting from a specified index.

    Args:
        large_string (str): The large string containing multiple paragraphs.
        dataframe (pd.DataFrame): The DataFrame where paragraphs will be inserted.
        start_index (int): The index in the DataFrame to start inserting paragraphs.

    Returns:
        pd.DataFrame: The updated DataFrame with paragraphs inserted.
    """
    # Split the large string into paragraphs (assuming paragraphs are separated by double newlines)
    paragraphs = large_string.split('\n\n')

    # Insert paragraphs into the DataFrame starting from the specified index
    size = len(paragraphs)
    for i, paragraph in enumerate(paragraphs):
        dataframe.at[start_index + i, 'prompt'] = paragraph.strip()

    return dataframe,start_index + size 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Load the data from the Excel file
file_path = 'Book1.csv'  # replace with your file path
df = pd.read_csv(file_path)

# Display a bar chart of the counts of each class in the intent column
intent_counts = df['intent'].value_counts()

plt.figure(figsize=(10, 6))
intent_counts.plot(kind='bar')
plt.xlabel('Intent Classes')
plt.ylabel('Count')
plt.title('Counts of Each Class in the Intent Column')
plt.show()
min_count = intent_counts.min()
```

```python
# Resample each class to the minimum count
balanced_df = pd.DataFrame()

for intent in intent_counts.index:
    class_subset = df[df['intent'] == intent]
    class_subset_resampled = resample(class_subset, 
                                      replace=False,  # sample without replacement
                                      n_samples=min_count,  # to match the least occurring class
                                      random_state=42)  # for reproducibility
    balanced_df = pd.concat([balanced_df, class_subset_resampled])
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python




# Display the counts of each class in the balanced dataframe
balanced_intent_counts = balanced_df['intent'].value_counts()

plt.figure(figsize=(10, 6))
balanced_intent_counts.plot(kind='bar')
plt.xlabel('Intent Classes')
plt.ylabel('Count')
plt.title('Counts of Each Class in the Balanced Intent Column')
plt.show()

# Save the balanced data to a new Excel file (optional)
balanced_df.to_csv('balanced_data.csv', index=False)
```

```python
import re
def replace_words_in_brackets(text):
    # Use a regular expression to find words encased in <>
    pattern = re.compile(r'<(.*?)>')
    
    # Replace each match with the word itself (removing the brackets)
    result = pattern.sub(r'\1', text)
    
    return result
```
